```python
import pandas as pd

# Load the raw dataset
df = pd.read_csv('../data/raw/WB_WDI_SN_ITK_DEFC_ZS.csv')

# Show the first 5 rows
df.head()
```

```python
# Check shape and basic info
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
print("\nColumn names:")
print(df.columns.tolist())
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Keep only the useful columns
df_clean = df[['REF_AREA_LABEL', 'TIME_PERIOD', 'OBS_VALUE']].copy()

# Rename columns to simpler names
df_clean.columns = ['Country', 'Year', 'Undernourishment_Pct']

# Check the result
df_clean.head()
```

```python
# Check what unique years we have
print("Years in dataset:")
print(sorted(df_clean['Year'].unique()))

print(f"\nTotal unique countries/regions: {df_clean['Country'].nunique()}")

# Preview some of the region names we need to remove
print("\nSample entries:")
print(df_clean['Country'].unique()[:30])
```

```python
# List of regional groupings to remove
regions_to_remove = [
    'World', 'Arab World', 'Caribbean small states',
    'Central Europe and the Baltics', 'Early-demographic dividend',
    'East Asia & Pacific', 'East Asia & Pacific (excluding high income)',
    'East Asia & Pacific (IDA & IBRD)', 'Euro area',
    'Europe & Central Asia', 'Europe & Central Asia (excluding high income)',
    'Europe & Central Asia (IDA & IBRD)', 'Eastern & Southern Africa',
    'Western & Central Africa', 'Sub-Saharan Africa (IDA & IBRD)',
    'Upper middle income', 'Lower middle income', 'Low income',
    'High income', 'Low & middle income', 'Middle income',
    'Fragile and conflict affected situations',
    'Heavily indebted poor countries (HIPC)',
    'Late-demographic dividend', 'Latin America & Caribbean',
    'Latin America & Caribbean (excluding high income)',
    'Latin America & the Caribbean (IDA & IBRD)',
    'Least developed countries: UN classification',
    'Middle East & North Africa',
    'Middle East & North Africa (excluding high income)',
    'Middle East & North Africa (IDA & IBRD)',
    'North America', 'OECD members', 'Other small states',
    'Pacific island small states', 'Pre-demographic dividend',
    'Small states', 'South Asia', 'South Asia (IDA & IBRD)',
    'Sub-Saharan Africa', 'Sub-Saharan Africa (excluding high income)',
    'Post-demographic dividend', 'IDA & IBRD total', 'IDA blend',
    'IDA only', 'IDA total', 'IBRD only', 'Not classified'
]

# Filter out regions
df_clean = df_clean[~df_clean['Country'].isin(regions_to_remove)]

print(f"Rows remaining: {len(df_clean)}")
print(f"Unique countries remaining: {df_clean['Country'].nunique()}")
```

```python
# Round the percentage to 2 decimal places
df_clean['Undernourishment_Pct'] = df_clean['Undernourishment_Pct'].round(2)

# Make sure Year is an integer
df_clean['Year'] = df_clean['Year'].astype(int)

# Reset the index
df_clean = df_clean.reset_index(drop=True)

# Final check
print(f"Final dataset shape: {df_clean.shape}")
print(f"\nYear range: {df_clean['Year'].min()} - {df_clean['Year'].max()}")
print(f"\nSample of cleaned data:")
df_clean.head(10)
```

```python
# Save the cleaned dataset
df_clean.to_csv('../data/cleaned/WB_undernourishment_cleaned.csv', index=False)

print("Cleaned dataset saved successfully!")
```
